```python
import pandas as pd
```

```python
df = pd.read_csv("../data/complaints.csv")
```

```python
df
```

```python
df['Consumer complaint narrative'] = df['Consumer complaint narrative'].str.lower()
df = df.rename(columns={'Consumer complaint narrative':'complaint_text'})
df
```

```python
import re

df['complaint_text'] = df['complaint_text'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))
df['complaint_text'] = df['complaint_text'].apply(lambda x: re.sub(r'\b[xX]{2,}\b', '', x))
```
